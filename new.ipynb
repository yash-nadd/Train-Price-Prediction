```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pickle

# Load and preprocess the data
data = pd.read_csv("Clean_Dataset.csv")
df1 = data.fillna(method="ffill")

# Encode categorical features
label_encoders = {}
for col in ['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']:
    le = LabelEncoder()
    df1[col] = le.fit_transform(df1[col])
    label_encoders[col] = le

# Drop unnecessary column and split data
df1.drop('flight', axis=1, inplace=True)
df1.drop('stops',axis = 1,inplace = True)
df1.drop('days_left',axis = 1,inplace = True)
df1.drop('S.No',axis = 1,inplace = True)
X = df1.drop('price', axis=1)
y = df1['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train the model
model = RandomForestRegressor(n_estimators=7)
model.fit(X_train, y_train)

# Save the model and label encoders
with open('model1.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)
with open('label_encoders.pkl', 'wb') as le_file:
    pickle.dump(label_encoders, le_file)

print("Model and encoders saved successfully.")
```
